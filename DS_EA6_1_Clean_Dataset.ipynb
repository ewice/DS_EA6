```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv')

# Komplett leere Zeilen entfernen
df.dropna(how='all', inplace=True)

# Die Spalten Id und FullName entfernen
# Zur Datensparsamkeit werden die beiden Spalten nicht benötigt
# Als Id kann der Index verwendet werden
# FullName kann aus FirstName und LastName zusammengesetzt werden
df.drop(['id', 'full_name'], inplace=True, axis=1)

# Alter korrigieren
# Age Type von String zu Numeric ändern und Fehler rausschmeißen
# Möglicher Tippfehler von -78 zu 78 umwandeln
df.age = pd.to_numeric(df.age, errors='coerce').abs()

# Nicht angegebenes Geschlecht auf "NotSpecified" ändern
# Geschlechtsangabe teilweise nicht benötigt und zur weiteren Verarbeitung 'NaN' genauer benannt
df.gender = df.gender.fillna('NotSpecified')

# Duplikate entfernen
df.drop_duplicates(inplace=True)

# Restliche NaN entfernen, da die nicht korrigierten Datensätze unvollständig sind
df.dropna(inplace=True)

# Index neu vergeben
df.reset_index(drop=True, inplace=True)

df
```
